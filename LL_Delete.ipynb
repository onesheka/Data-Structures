### Deleting Values
1. First Element
2. Last Element
3. In between <br>
   3.1  By value <br>
   3.2  By Node <br>
   3.3  By position <br>
   

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
    def get_next(self):
        return self.next
```

```python
class Linked_list:
    def __init__(self):
        self.head = None
    def traverse_LL(self):
        current = self.head
        while(current!=None):
            print(current.data)
            current = current.get_next()
        
    def length_LL(self):
        current = self.head
        count = 0
        while(current!=None):
            count+=1
            current = current.get_next()
        return count
    def del_beg(self):   ### deletion at beginning
        current = self.head
        self.head = current.get_next()
        
    def del_end(self):   ### deletion at end
        
        prev_node = self.head
        current = prev_node.next
        
        if(self.length_LL()==1):
            self.head = None
            print("Empty Linked list")
        else:
            while(current.get_next()!=None):
                    prev_node = prev_node.get_next()
                    current = current.get_next()
            prev_node.next = None
            
       
        
    def del_val(self,val):   ### deletion by value
        prev_node = self.head
        current = prev_node.get_next()
        if(prev_node.data==val):
            self.head = prev_node.get_next()
        else:
            while(current!=None):
                if(current.data==val):
                    prev_node.next = current.get_next()
                current = current.get_next()
                prev_node = prev_node.get_next()
                
    def del_node(self,node):   ### deleting values by nodes
        prev_node = self.head
        current = prev_node.next
        if(node==prev_node):
            self.head = prev_node.get_next()
        else:
            while(current!=None):
                if(current==node):
                    prev_node.next = current.next
                current = current.get_next()
                prev_node = prev_node.get_next()
                
    def del_pos(self,pos):   ### deleting values by position
        prev_node = self.head
        current = prev_node.get_next()
        n = 0
        if(pos==0):
            self.head = prev_node.get_next()
        else:
            while(current!=None):
                if(n+1==pos):
                    prev_node.next = current.next
                n = n + 1
                prev_node = prev_node.get_next()
                current = current.get_next()       
```
